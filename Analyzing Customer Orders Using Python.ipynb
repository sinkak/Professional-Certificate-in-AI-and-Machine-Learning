```python
from collections import defaultdict

#Create a list of customer names
customer_names = ["Abby","Bob","Sam","Pam","Ram"]
#Store each customer's order details (customer name, product, price, category) as tuples inside a list
customers_order_details = [("Abby","Cellphone","400","Electronics"),
                           ("Bob","Shirts","80","Clothing"),
                           ("Bob","fan","190","Electronics"),
                           ("Sam","Carpets","100","Essentials"),
                           ("Pam","Dresses","30","Clothing"),
                           ("Pam","freshener","20","Essentials"),
                           ("Pam","Dresses","30","Clothing"),
                           ("Ram","Curtains","200","Essentials"),
                           ("Ram","Dresses","50","Clothing")]
#TASK 1
#Write a dictionary where keys are customer names and values are lists of ordered products
# Create a defaultdict with a default value of a list
customer_orders = defaultdict(list)
# Iterate through the list  and append to the defaultdict
for name,product,price,category in customers_order_details:
    customer_orders[name].append(product)
# Convert back to a regular dict for final output if needed
customer_orders_dict = dict(customer_orders)
print("1. Customer order details")
print(customer_orders_dict)
print("-----------------------------")

#TASK 2
#Use a dictionary to map each product to its respective category
categories = defaultdict(list)
for name,product,price,category in customers_order_details:
    categories[category].append(product)
#Create a set of unique product categories
categories_dict = dict(categories)
print("2. Available Categories")
print(categories_dict)
#Display all available product categories
print("2.1 Available Product Categories")
print(list(categories_dict.keys()))
print("-----------------------------")


#TASK 3
# • Use a loop to calculate the total amount each customer spends
# • If the total purchase value is above $100, classify the customer as a high-value buyer
# • If it is between $50 and $100, classify the customer as a moderate buyer
# • If it is below $50, classify them as a low-value buyer
total_amount = defaultdict(int) #initialize a defaultdict with int as the default factory
for name,product,price,category in customers_order_details:
    #total_amount[name].append(int(price))
    total_amount[name]+=int(price)
# Convert back to a regular dict for final output if needed
prices_dict = dict(total_amount)
print("3. Total amount each customer spent")
print(prices_dict)


for key,value in prices_dict.items():
    if value > 100:
        print(key+" is a High-value buyer")
    elif 50 < value <= 100:
        print(key+" is a Moderate value buyer")
    else:
        print(key+" is a Low-value buyer")
        
print("-----------------------------")


#  Task 4 Generate business insights
# • Calculate the total revenue per product category and store it in a dictionary
# • Extract unique products from all orders using a set
# • Use a list comprehension to find all customers who purchased electronics
# • Identify the top three highest-spending customers using sorting
total_amount_product = defaultdict(int)
for name,product,price,category in customers_order_details:
        total_amount_product[product]+=int(price)
total_amount_product_dict = dict(total_amount_product)
print("4. Total amount spent on each product")
print(total_amount_product_dict)  
print("4.1 unique products that are shopped")
unique_products=[]
for name,product,price,category in customers_order_details:
    unique_products.append(product)
print(set(unique_products)) #Extract unique products from all orders using a set
print("4.2 Customers who purchased electronics")
name = [name for name,product,price,category in customers_order_details if category == "Electronics"] #Use a list comprehension to find all customers who purchased electronics
print(name)
print("4.3 Top three highest-spending customers")
top3_prices_dict_sorted = {k: v for k, v in sorted(prices_dict.items(), key=lambda item: item[1], reverse=True)[:3]} #sort a dictionary in a descending order based on price and print top 3 customers
print(top3_prices_dict_sorted)

print("-----------------------------")



# Task 5. Organize and display data
# • Print a summary of each customer’s total spending and their classification
# • Use set operations to find customers who purchased from multiple categories
# • Identify common customers who bought both electronics and clothing
customer_total_spending = defaultdict(lambda: [0, set()])
for name,product,price,category in customers_order_details:
    customer_total_spending[name][0]+=int(price)
    customer_total_spending[name][1].add(category)
    
customer_total_spending_dict = dict(customer_total_spending)
print("5.1 customer’s total spending and their categories")
print(customer_total_spending_dict)
print("5.2 Customers who purchased from multiple categories")
for key,value in customer_total_spending_dict.items():
    if len(value[1])>1:
        print(key)
print("5.3 Common customers who bought both electronics and clothing")
strings_to_find = ["Electronics", "clothing"]
def is_subset_ignore_case(list1, list2):
    # Convert all elements in both lists to lowercase for case-insensitive comparison
    lower_list1 = [x.lower() for x in list1]
    lower_list2 = [x.lower() for x in list2]
    
    # Check if all elements of lower_list1 are present in lower_list2
    return all(item in lower_list2 for item in lower_list1)

for key,value in customer_total_spending_dict.items():
    if is_subset_ignore_case(strings_to_find, value[1]):
        print(f"Name : {key}")

print("-----------------------------")
```
